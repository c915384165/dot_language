```python
from graphviz import Digraph
```

```python
dot = Digraph(engine='neato')

dot.attr(label="test graph label",rankdir="LR")
print(dot.source)
```

```python
dot = Digraph('a',engine='neato',format='svg')
dot.graph_attr={'shape':"record", "color":'red'}
dot.node_attr={'shape':"box","color":"white"}
dot.edge('a', 'b')
dot.node_attr={'shape':"box","color":"black"}
dot.edge('c','d')
dot.node_attr["shape"]="ellipse"
dot.node('a','about',color="red")
print(dot.source)
```

```python
dot.graph_attr["randir"] = "LR"
```
