```python
#数据读取
# 如果是空行怎么办？
def load_file_txt(file_name):
    datals = []
    f = open(file_name)
    for line in f:
        line = line.replace("\n","")
#         datals.append(list(map(eval, line.split(","))))
        datals.append(list(line.split(";")))
    f.close()
    return datals
```

```python
def add_tag_si (s,i=''):
    return('<'+s+str(i)+'>')


print(add_tag_si('',3))


def add_tag_for_list(l):
    length = len(l)
    ls=[]
    for i in range(length):
        ls.append(add_tag_si('',i)+' '+str(l[i]))
    return('|'.join(ls))

ls=['a','b','c']
add_tag_for_list(ls)
```
