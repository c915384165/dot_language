node_attr的三种用法。记住！是node，不是nord也不是nerd。
node_attr是一个字典变量，可以快速赋予node节点属性。这样就不用在生成每个node时候单独设置。

使用node_attr有一下三种方法：
1. 在实例化对象时候，对象参数中添加 record=Digraph(node_attr={"shape":"record"})
2. 在创建对象后添加，这样的好处是可以为后边所有的node确定属性。
3. 第三种最普通，就是在创建节点的时候在函数中添加变量。注意！这里添加时候是以单独变量添加，而不是词典。shape=“record”

```python
from graphviz import Digraph

record = Digraph(comment="a new Digraph for python3", node_attr={"shape":"record"})

record.node('node0','<f0> a|<f1> b|<f2> c')

record.node('node1', '<f0> d|<f1> e|<f2> f')

record.edge("node0:f0", "node1:f1")

record.edges([("node0:f1","node1:f0")])

record
```

```python
from graphviz import Digraph

record = Digraph(comment="a new Digraph for python3")
record.node_attr["shape"]="record"

record.node('node0','<f0> a|<f1> b|<f2> c')

record.node('node1', '<f0> d|<f1> e|<f2> f')

record.edge("node0:f0", "node1:f1",color="blue")

record.edges([("node0:f1","node1:f0")])

record
```

```python
def add_tag_for_dic(d):
    ls = []
    for i in d.keys():
        ls.append(add_tag_si(i)+' '+str(d[i]))
    return('|'.join(ls))
```

```python
l1 = [1,3,4,5,6]
l2 = ['a','b','c','d']
```

```python
def add_tag_for_list(l):
    length = len(l)
    ls=[]
    for i in range(length):
        ls.append(add_tag_si('',i)+' '+str(l[i]))
    return('|'.join(ls))
```

```python
add_tag_for_list(l1)
```
